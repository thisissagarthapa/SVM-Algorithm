# Support Vector Machine Algorithm

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset=pd.read_csv("school_data.csv")
```

```python
dataset.isnull().sum()
```

```python
dataset.head(3)
```

```python
sns.scatterplot(x="Exam Score",y="Attendance",hue="Pass/Fail",data=dataset)
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Pass/Fail"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.2,random_state=40)
```

```python
from sklearn.svm import SVC
```

```python
sv=SVC(kernel="linear")
```

```python
# linear,poly,rbf,sigmoid
```

```python
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100
```

```python
sv.score(x_train,y_train)*100
```

```python
sv.predict([[81,169]])
```

```python
sv.predict([[44,95]])
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv)
plt.show()
```
